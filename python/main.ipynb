```python
import pandas as pd
```

```python
df = pd.read_excel('dados.xlsx')
```

```python
df.head()
```

```python
df.columns
```

```python
colunas = [
    'data',
    'faixa_etaria',
    'frequencia_que_leva_pet',
    'avaliacao_atendimento',
    'avaliacao_marcar_hora',
    'demonstracao_cuidado',
    'pos_atendimento',
    'recebeu_mensagem_promocionais',
    'porque_continuar_cliente',
    'melhorias_empresa',
    'nota_empresa',
    'comentarios_adicionais'
]
df.rename(columns={old: new for old, new in zip(df.columns, colunas)}, inplace=True)
```

```python
df.columns
```

```python
len(df['avaliacao_atendimento'])
```

```python
def pegar_num(x):
    for i in range(len(x)):
        return (x[i][0])
df['avaliacao_atendimento'] = df['avaliacao_atendimento'].apply(pegar_num).astype(int)
df['avaliacao_marcar_hora'] = df['avaliacao_marcar_hora'].apply(pegar_num).astype(int)
df['demonstracao_cuidado'] = df['demonstracao_cuidado'].apply(pegar_num).astype(int)
```

```python
def pegar_num(x):
    for i in range(len(x)):
        if x[i][0] == 'A':
            return 0
        else:
            return (x[i][0])
df['pos_atendimento'] = df['pos_atendimento'].apply(pegar_num).astype(int)
```

```python
df['pos_atendimento'].value_counts()
```

```python
df['comentarios_adicionais'].dtype
```

```python
df = df.fillna('Nao')
```

```python
df.head()
```

```python
df['faixa_etaria'] = df['faixa_etaria'].str.capitalize()
```

```python
df['data'] = df['data'].dt.strftime('%d/%m/%Y')
```

```python
df.to_csv('dados_limpos.csv', index=False)
```
